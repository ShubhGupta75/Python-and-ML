```python
import seaborn as sns
```

```python
# Basic Plotting

tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill', y='tip', data=tips)
plt.title('Scatter plot for total bill vs tip')
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical Plots
## Bar Plot

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
## Box Plot => A boxplot summarizes data using quartiles and highlights outliers. Minimum, Q1 (25th percentile), Median (Q2), Q3 (75th percentile) and Maximum.

sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
## Violin plot => A violin plot adds density information, showing the full distribution shape. Width of violin â†’ Data density, Thick area â†’ High concentration, Thin area â†’ Low concentration, Often includes median & quartiles inside

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
### Histograms

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## KDE Plot => (Kernel Density Estimation Plot), A histogram groups them into bins. A KDE plot estimates the underlying distribution smoothly using kernels

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pair plot => shows relationship between every variable(features)

sns.pairplot(tips)
```

```python
## Heatmap

corr=tips[['total_bill','tip','size']].corr()
corr

# Correlation measures linear relationship strength between two variables.
# The matrix shows moderate to strong positive relationships among total_bill, tip, and size. Higher bills and larger group sizes generally lead to higher tips.
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')\

# total_bill & tip â†’ 0.6757(Moderately strong positive correlation)
# As bill increases, tip generally increases
```

```python
import pandas as pd

sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## Plot total sales by product

plt.figure(figsize=(10,6))
sns.barplot(x = 'Product Category', y='Total Revenue', data = sales_df, estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show

# estimator defines which aggregation function is applied to the y values for each category in x.
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = 'Region', y='Total Revenue', data = sales_df, estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show
```
