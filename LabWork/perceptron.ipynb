```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def train_perceptron(X, Y, lr=0.5, epochs=20):

    # Initial parameters
    w = np.array([0.5, 0.6], dtype=float)
    b = 0.3

    losses = []

    for epoch in range(epochs):
        total_loss = 0

        for i in range(len(X)):

            # Forward pass
            z = np.dot(w, X[i]) + b
            y_hat = sigmoid(z)

            # Loss (Least Squares Error)
            error = 0.5 * (Y[i] - y_hat)**2
            total_loss += error

            # Backpropagation
            delta = (Y[i] - y_hat) * y_hat * (1 - y_hat)

            # Update weights
            w += lr * delta * X[i]
            b += lr * delta

        losses.append(total_loss)

        print(f"Epoch {epoch+1}: Loss = {total_loss:.4f}")

    return w, b, losses
```

```python
X1 = np.array([[0,0],
               [0,1],
               [1,0],
               [1,1]])

Y1 = np.array([0,1,1,1])

w1, b1, loss1 = train_perceptron(X1, Y1)
```

```python
X2 = np.array([[0,0],
               [0,1],
               [1,0],
               [1,1]])

Y2 = np.array([0,0,0,1])

w2, b2, loss2 = train_perceptron(X2, Y2)
```

```python
X3 = np.array([[0,0],
               [0.5,0.5],
               [1.3,0.5],
               [0.8,1.2]])

Y3 = np.array([0,1.1,1.2,1.0])

w3, b3, loss3 = train_perceptron(X3, Y3)
```

```python
plt.figure()
plt.plot(loss1)
plt.title("Dataset 1 (OR) - Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

plt.figure()
plt.plot(loss2)
plt.title("Dataset 2 (AND) - Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

plt.figure()
plt.plot(loss3)
plt.title("Dataset 3 (Continuous) - Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
def plot_decision_boundary(X, Y, w, b, title):

    plt.figure()

    # Scatter points
    plt.scatter(X[:,0], X[:,1])

    # Decision boundary line
    x_vals = np.linspace(-0.5, 1.5, 100)
    y_vals = -(w[0]*x_vals + b)/w[1]

    plt.plot(x_vals, y_vals)
    plt.title(title)
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
```

```python
plot_decision_boundary(X1, Y1, w1, b1, "Dataset 1 Decision Boundary")
```

```python
plot_decision_boundary(X2, Y2, w2, b2, "Dataset 2 Decision Boundary")
```

```python
def print_predictions(X, Y, w, b):

    print("\nPredicted vs Actual\n")

    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_hat = sigmoid(z)
        print(f"Input: {X[i]} | Predicted: {y_hat:.4f} | Actual: {Y[i]}")
```

```python
print_predictions(X1, Y1, w1, b1)
print_predictions(X2, Y2, w2, b2)
print_predictions(X3, Y3, w3, b3)
```
