### Introduction to Tuples

Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
# Conversion of list to tuple

numbers=tuple([1,2,3,4,5,6])
numbers
# list((1,2,3,4,5,6))
```

##### Indexing and Slicing operation is same as List

```python
numbers
```

```python
mixed_tuple=(1,"Hello World",3.14, True)
```

```python
## Tuple Operations

concatenation_tuple=numbers + mixed_tuple
print(concatenation_tuple)
```

```python
numbers *3
```

```python
## Immutable Nature Of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1]="Shubh"
print(lst)

# numbers[1]="Krish"  => not possible
```

```python
## Tuple Methods
print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and Unpacking tuple
## packing => if terms are written in comma seperarted manner, they are packed into tuple.
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
##Unpacking a tuple

a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *

numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)  # it will print all numbers after first and before last.
print(last)
```

```python
## Nested Tuple
## Nested List

lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")  ## end = " "  =>  to print in same line
    print()
```
