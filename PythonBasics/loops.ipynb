```python
for i in range(1,10,2):
  print(i)

# in range(), 1st term is the starting point, 2nd term is the ending point, and 3rd term is step size
```

```python
## Nested loopss
## a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")

# When i is running 1 time, j is running 2 times
# The f before the string indicates an f-string.
# An f-string allows you to embed variables or expressions directly inside a string using {}.
# Similar to String interpolation in JS.
```

```python
# Sum of first n natural number

n = int(input("Enter a number"))
sum = 0
count = 1
while (count<=n):
  sum = sum + count
  count = count + 1

print(sum)
```

```python

# Display all prime numbers till 100

for num in range(1,101):
  if num > 1:
    for i in range(2, num):
      if num % i == 0:
        break
    else:
      print(num)
```
