SMOTE (Synthetic Minority Over-sampling Technique) is a method used in machine learning to handle imbalanced datasets by creating synthetic samples for the minority class instead of simply duplicating existing ones.

How SMOTE works

Instead of copying minority samples, SMOTE:

1. Selects a minority class sample.

2. Finds its k nearest neighbors (usually k=5).

3. Generates new synthetic points between the sample and its neighbors.

This creates new realistic data points.

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_redundant=0, n_features=2, n_clusters_per_class=1,
                    weights=[0.9], random_state=12)
# weights means one class will be 90% more than other class.
```

```python
import pandas as pd

df1 = pd.DataFrame(X, columns=['f1', 'f2'])
df2 = pd.DataFrame(y, columns=['target'])
final_df = pd.concat([df1, df2], axis=1)
final_df.head()
```

```python
final_df.target.value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
# c => assigns different colors to points based on the target values.
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# Transform the dataset

oversample = SMOTE()
X,y = oversample.fit_resample(final_df[['f1','f2']], final_df['target'])
```
