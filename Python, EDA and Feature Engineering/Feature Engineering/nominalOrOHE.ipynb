#### Nominal/OHE Encoding
One hot encoding, also known as nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green, blue), we can represent it using one hot encoding as follows:

1. Red: [1, 0, 0]
2. Green: [0, 1, 0]
3. Blue: [0, 0, 1]

### Disadvantages

1. If a feature has many categories, OHE creates many new columns. (If a feature has many categories, OHE creates many new columns.)

2. Most values in the new columns are 0, with only one 1 per row. This creates sparse matrices, which can slow down some algorithms and it leads to overfiting.


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
  'color': ['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
# Create an instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
# Perform fit and transform
# fit_transform() learns from the data and immediately applies the transformation to it.
# In One-Hot Encoding, fit_transform() first learns all categories in the column and then converts them into binary columns at the same time.

encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
# For new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df], axis=1)
```
