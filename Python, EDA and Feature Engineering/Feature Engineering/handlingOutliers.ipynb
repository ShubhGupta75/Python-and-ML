### 5 number summary and Box plot 

```python
import numpy as np
```

```python
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum, Q1, median, Q3, maximun = np.quantile(lst_marks, [0,0.25,0.5,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximun

# This code calculates five important statistical values (quartiles) from the list of marks using numpy.
# np.quantile() divides the data into equal parts based on percentage positions.
```

```python
# IQR measures the spread of the middle 50% of the data.
# IQR is commonly used to detect outliers.

IQR = Q3 -Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5*IQR
higher_fence = Q3 + 1.5*IQR
```

```python
higher_fence  # anything higher than 141.5 will be considered as outlier
```
